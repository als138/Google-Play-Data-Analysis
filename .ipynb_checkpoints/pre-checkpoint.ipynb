```python
import pandas as pd
import numpy as np

# Load the dataset
df = pd.read_csv("Google-Playstore.csv")

# Remove duplicate records
df.drop_duplicates(inplace=True)

# 1. App Name: Remove extra spaces
df['App Name'] = df['App Name'].astype(str).str.strip()

# 2. App Id: Remove extra spaces
df['App Id'] = df['App Id'].astype(str).str.strip()

# 3. Category: Remove extra spaces
df['Category'] = df['Category'].astype(str).str.strip()

# 4. Rating: Convert to numeric and validate the range (should be between 0 and 5)
df['Rating'] = pd.to_numeric(df['Rating'], errors='coerce')
df = df[(df['Rating'].isnull()) | ((df['Rating'] >= 0) & (df['Rating'] <= 5))]
df.dropna(subset=['Rating'], inplace=True)

# 5. Rating Count: Convert to integer
df['Rating Count'] = pd.to_numeric(df['Rating Count'], errors='coerce')
df.dropna(subset=['Rating Count'], inplace=True)

# 6. Installs: Remove commas and '+' symbol, then convert to integer
df['Installs'] = df['Installs'].astype(str)\
    .str.replace(',', '', regex=False)\
    .str.replace('+', '', regex=False)
df['Installs'] = pd.to_numeric(df['Installs'], errors='coerce')
df.dropna(subset=['Installs'], inplace=True)

# 7. Minimum Installs: Remove commas and '+' symbol, then convert to integer
df['Minimum Installs'] = df['Minimum Installs'].astype(str)\
    .str.replace(',', '', regex=False)\
    .str.replace('+', '', regex=False)
df['Minimum Installs'] = pd.to_numeric(df['Minimum Installs'], errors='coerce')

# 8. Maximum Installs: Remove commas and '+' symbol, then convert to integer
df['Maximum Installs'] = df['Maximum Installs'].astype(str)\
    .str.replace(',', '', regex=False)\
    .str.replace('+', '', regex=False)
df['Maximum Installs'] = pd.to_numeric(df['Maximum Installs'], errors='coerce')

# 9. Free: Convert to boolean (True/False)
# If values are in string format, remove extra spaces and convert to lowercase
df['Free'] = df['Free'].astype(str).str.strip().str.lower()
df['Free'] = df['Free'].apply(lambda x: True if x in ['true', 'yes', 'free', '1'] else False)

# 10. Price: Remove dollar sign (or other currency symbols) and convert to numeric
df['Price'] = df['Price'].astype(str).str.replace('$', '', regex=False).str.strip()
df['Price'] = pd.to_numeric(df['Price'], errors='coerce')
df['Price'] = df['Price'].fillna(0)

# 11. Currency: Remove extra spaces
df['Currency'] = df['Currency'].astype(str).str.strip()

# 12. Size: Standardize app size (convert 'M' to megabytes and 'k' to megabytes)
def convert_size(size):
    if isinstance(size, str):
        size = size.strip()
        if size.lower() in ['varies with device', 'n/a', '']:
            return np.nan
        elif size.endswith('M'):
            try:
                return float(size[:-1])
            except:
                return np.nan
        elif size.endswith('k'):
            try:
                # Convert kilobytes to megabytes
                return float(size[:-1]) / 1024
            except:
                return np.nan
    return np.nan

df['Size'] = df['Size'].apply(convert_size)

# 13. Minimum Android: Remove extra spaces
df['Minimum Android'] = df['Minimum Android'].astype(str).str.strip()

# 14. Developer Id: Remove extra spaces
df['Developer Id'] = df['Developer Id'].astype(str).str.strip()

# 15. Developer Website: Remove extra spaces
df['Developer Website'] = df['Developer Website'].astype(str).str.strip()

# 16. Developer Email: Remove extra spaces
df['Developer Email'] = df['Developer Email'].astype(str).str.strip()

# 17. Released: Convert to date format
df['Released'] = pd.to_datetime(df['Released'], errors='coerce')
df.dropna(subset=['Released'], inplace=True)

# 18. Privacy Policy: Remove extra spaces
df['Privacy Policy'] = df['Privacy Policy'].astype(str).str.strip()

# 19. Last Updated: Convert to date format
df['Last Updated'] = pd.to_datetime(df['Last Updated'], errors='coerce')
df.dropna(subset=['Last Updated'], inplace=True)

# 20. Content Rating: Remove extra spaces
df['Content Rating'] = df['Content Rating'].astype(str).str.strip()

# 21. Ad Supported: Convert to boolean (True/False)
df['Ad Supported'] = df['Ad Supported'].astype(str).str.strip().str.lower()
df['Ad Supported'] = df['Ad Supported'].apply(lambda x: True if x in ['true', 'yes', '1'] else False)

# # 22. In-app purchases: Convert to boolean (True/False)
# df['In app purchases'] = df['In app purchases'].astype(str).str.strip().str.lower()
# df['In app purchases'] = df['In app purchases'].apply(lambda x: True if x in ['true', 'yes', '1'] else False)

# # 23. Editor's Choice: Convert to boolean (True/False)
# df['Editor Choice'] = df['Editor Choice'].astype(str).str.strip().str.lower()
# df['Editor Choice'] = df['Editor Choice'].apply(lambda x: True if x in ['true', 'yes', '1'] else False)

# Finally, remove records with NaN values in essential columns
essential_cols = ['App Name', 'App Id', 'Category', 'Rating', 'Rating Count', 'Installs', 'Price', 'Size', 'Released', 'Last Updated']
df.dropna(subset=essential_cols, inplace=True)

# Save the cleaned dataset to a new CSV file
df.to_csv("cleaned_googleplaystore.csv", index=False)

# Display a summary of the cleaned dataset
print("Dataset information after cleaning:")
print(df.info())
```
