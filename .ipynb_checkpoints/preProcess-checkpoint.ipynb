```python

from unidecode import unidecode

df['App Name'] = df['App Name'].apply(lambda x: unidecode(x) if isinstance(x, str) else x)
df['App Id'] = df['App Id'].apply(lambda x: unidecode(x) if isinstance(x, str) else x)
df['Category'] = df['Category'].apply(lambda x: unidecode(x) if isinstance(x, str) else x)
df.to_csv("Google-Playstore.csv", index=False)
```

```python
print(df.dtypes)
```

```python

df[['App Name', 'Category', 'App Id']] = df[['App Name', 'Category', 'App Id']].astype(str)
df.to_csv("Google-Playstore.csv", index=False)
```

```python
df['Installs'] = df['Installs'].str.replace("+", "", regex=False)
df.to_csv("Google-Playstore.csv", index=False)
```

```python
df['Minimum Android'] = df['Minimum Android'].str.replace("and up", "", regex=False)
df.to_csv("Google-Playstore.csv", index=False)
```
