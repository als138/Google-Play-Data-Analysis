```python
import pandas as pd
import psycopg2

# PostgreSQL Connection Settings
DB_NAME = "googleplay"
DB_USER = "postgres"
DB_PASSWORD = "1237"
DB_HOST = "localhost"
DB_PORT = "5433"

# Read the cleaned CSV file
df = pd.read_csv("cleaned_googleplaystore.csv")

# Connect to the database
conn = psycopg2.connect(
    dbname=DB_NAME, user=DB_USER, password=DB_PASSWORD, host=DB_HOST, port=DB_PORT
)
cursor = conn.cursor()

# 1️⃣ Load Categories
categories = df["Category"].dropna().unique()
category_map = {}
for category in categories:
    cursor.execute("INSERT INTO categories (name) VALUES (%s) ON CONFLICT (name) DO NOTHING RETURNING id;", (category,))
    category_id = cursor.fetchone()
    if category_id:
        category_map[category] = category_id[0]

# 2️⃣ Load Developers
developers = df[["Developer Id", "Developer Website", "Developer Email"]].dropna().drop_duplicates()
developer_map = {}
for _, row in developers.iterrows():
    cursor.execute("""
        INSERT INTO developers (developer_identifier, website, email) 
        VALUES (%s, %s, %s) 
        ON CONFLICT (developer_identifier) DO NOTHING RETURNING id;
    """, (row["Developer Id"], row["Developer Website"], row["Developer Email"]))
    
    developer_id = cursor.fetchone()
    if developer_id:
        developer_map[row["Developer Id"]] = developer_id[0]

# 3️⃣ Load Apps
for _, row in df.iterrows():
    category_id = category_map.get(row["Category"])
    developer_id = developer_map.get(row["Developer Id"])

    if not category_id or not developer_id:
        continue  # Skip apps with missing category or developer

    cursor.execute("""
        INSERT INTO apps (
            app_name, app_id, category_id, rating, rating_count, installs, minimum_installs, maximum_installs, free,
            price, currency, size, minimum_android, released, privacy_policy, last_updated, content_rating, 
            ad_supported, in_app_purchases, editor_choice, developer_id
        ) VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s)
        ON CONFLICT (app_id) DO NOTHING;
    """, (
        row["App Name"], row["App Id"], category_id, row["Rating"], row["Rating Count"], row["Installs"],
        row["Minimum Installs"], row["Maximum Installs"], row["Free"], row["Price"], row["Currency"], 
        row["Size"], row["Minimum Android"], row["Released"], row["Privacy Policy"], row["Last Updated"],
        row["Content Rating"], row["Ad Supported"], row["In App Purchases"], row["Editors Choice"], developer_id
    ))

# Commit changes and close connection
conn.commit()
cursor.close()
conn.close()
```
